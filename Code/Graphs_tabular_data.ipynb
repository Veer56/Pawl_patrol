# file with seaborn plots of Tabular data from petfinder.my

In this notebook is a simple overview of the Tabular data in displots. This gives a indication if there is any correlation between a feature and the pawpularity.

```python
'''Import the libraries'''
import os
import cv2
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# csv file as a dataframe
url = 'https://raw.githubusercontent.com/Veer56/Pawl_patrol/main/Code/data/train.csv'
train_csv = pd.read_csv(url)
display(train_csv.head(5))
display(train_csv.describe())
```

```python
# plot barchart of the amount of pictures for every pawpularity score
train_csv
sns.displot(train_csv, x="Pawpularity")
```

```python
''' Make a list of the columnnames without pawpularity and Id''' 
list_columns = list(train_csv.columns.values)
list_columns.remove('Pawpularity')
list_columns.remove('Id')

# plot every feature from the list in a displot
for name in list_columns:
    sns.displot(train_csv, x="Pawpularity", hue = name, kde=True)
    
```
