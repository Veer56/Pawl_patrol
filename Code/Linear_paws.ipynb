```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# read the data
data = pd.read_csv('data/train.csv')
data_test = pd.read_csv('data/test.csv')

# look at what is there
display(data.head(10))
display(data_test.head(10))
```

```python
# get the y and x values
y = data['Pawpularity']
X = data.drop(['Id', 'Pawpularity'], axis=1)
X_test = data_test.drop(['Id'], axis=1)


def addones(X):
    """Add a row of ones to the X matrix"""
    iota = np.ones((len(X), 1))
    X = np.hstack((iota, X))
    return X

X = addones(X)
X_test = addones(X_test)
```

```python
def norm(X,y):
    """calculate the linear coefficients with X an y"""
    b = np.linalg.inv(X.T @ X) @ X.T @ y
    return b

b = norm(X,y)
print(b)
```

```python
def costs(X, y, b):
    """Get the costs"""
    e = y - X @ b
    cost = 1 / len(y) * e.T @ e
    return cost
```

```python
cost = costs(X,y,b) 
print(cost)
```

```python
# Get the y_test values
y_test = X_test @ b
print(y_test)
```
